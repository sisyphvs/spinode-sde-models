```python
import numpy as np
from dynamics import stoch_dyn_BS

dt = 1/252
num_trajectories = 10000
num_inits = 50
T = 100

nx = 1
nw = 1

S0_vals = np.linspace(50, 150, num_inits)
mean_i, mean_f, cov_i, cov_f = [], [], [], []

for S0 in S0_vals:
    X, X_next = [], []
    for _ in range(num_trajectories):
        Sk = S0
        for t in range(T):
            Skw = np.random.randn()
            [Skp1, _, _] = stoch_dyn_BS([Sk, Skw])
            if t == T-2:
                X.append(Sk)
                X_next.append(Skp1)
            Sk = Skp1
    X, X_next = np.array(X), np.array(X_next)
    mean_i.append([np.mean(X)])
    mean_f.append([np.mean(X_next)])
    cov_i.append([[np.var(X)]])
    cov_f.append([[np.var(X_next)]])

np.save("BS/data/mean_initial.npy", np.array(mean_i))
np.save("BS/data/mean_final.npy", np.array(mean_f))
np.save("BS/data/cov_initial.npy", np.array(cov_i))
np.save("BS/data/cov_final.npy", np.array(cov_f))
```
