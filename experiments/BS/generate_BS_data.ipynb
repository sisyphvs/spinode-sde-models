```python
import os
import sys
import numpy as np

# Ruta al directorio raíz del repo (dos niveles hacia arriba desde experiments/BS)
repo_root = os.path.abspath(os.path.join(os.getcwd(), "../.."))

# Agregar el root al sys.path si no está
if repo_root not in sys.path:
    sys.path.append(repo_root)

from src.dynamics import stoch_dyn_BS
```

```python
dt = 1/252
num_trajectories = 10000
num_inits = 50
T = 100

nx = 1
nw = 1

S0_vals = np.linspace(50, 150, num_inits)
mean_i, mean_f, cov_i, cov_f = [], [], [], []

for S0 in S0_vals:
    X, X_next = [], []
    for _ in range(num_trajectories):
        Sk = S0
        for t in range(T):
            Skw = np.random.randn()
            [Skp1, _, _] = stoch_dyn_BS([Sk, Skw])
            if t == T-2:
                X.append(Sk)
                X_next.append(Skp1)
            Sk = Skp1
    X, X_next = np.array(X), np.array(X_next)
    mean_i.append([np.mean(X)])
    mean_f.append([np.mean(X_next)])
    cov_i.append([[np.var(X)]])
    cov_f.append([[np.var(X_next)]])
```

```python
np.save("data/mean_initial.npy", np.array(mean_i))
np.save("data/mean_final.npy", np.array(mean_f))
np.save("data/cov_initial.npy", np.array(cov_i))
np.save("data/cov_final.npy", np.array(cov_f))
```
